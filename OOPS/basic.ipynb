```python
class Student:
  __slots__ = ("_Student__name", "age") # give variable names as in which to use
  def __init__(self,name,age):
    self.__name = name
    self.age = age
  def isTeen(self):
    return self.age > 16

s1 = Student("ad",17)
s1.isTeen()

```

```python
class Vechile:
  def __init__(self,ms,color):
    self.__maxspeed = ms
    self.color = color
  def getMaxSpeed(self):
    return self.__maxspeed
  def print(self):
    print("Vechile Print")

class Car(Vechile):
  def __init__(self, ms, color,type):
    super().__init__(ms,color)
    self.type = type
  def printCar(self):
    self.print() # Print method is first searched in the child if not found then taken to Parent class for search
    super().print()
    print(f"Type: {self.type}, Color: {self.color}, Max Speed: {self.getMaxSpeed()}")
```

```python
c1 = Car(100,"red","suv")
c1.printCar()
```

```python
class Father():
  def __init__(self):
    self.name = "Father"
  def print():
    print("Inside Father class")

class Mother():
  def __init__(self):
    self.name = "Mother"

  @staticmethod
  def print():

    print("Inside Mother class")

  @classmethod
  def returnM(cls):
    return cls

m= Mother()
m.print()
m1 = Mother.returnM()
m1.print()
```

```python
class son(Mother,Father):
  def __init__(self):
    super().__init__()
    super().print()
```
